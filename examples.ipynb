```python
# need to improt pandas
from matplotlib import pyplot as plt 
```

```python
x=[1,3,5,7]
y=[1,2,3,4]
plt.plot(x,y)
plt.title('test plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x=[1,3,5,7]
y=[1,2,3,4]
z=[9,5,1,3]
plt.plot(x,y)
plt.plot(x,z)
plt.title('test plot')
plt.xlabel('x')
plt.ylabel('y' and 'z')
plt.show()
```

```python
# legend arguments :
x=[1,3,5,7]
y=[1,2,3,4]
z=[9,5,1,3]
plt.plot(x,y)
plt.plot(x,z)
plt.title('test plot')
plt.xlabel('x')
plt.ylabel('y' and 'z')
plt.legend(['this is y','this is z'])
plt.show()
```

```python
data = pd.read_csv("countries.csv")
data
```

```python
# here we are going to do comapre of the population of us and china
us = data[data.country =='United States']
us
```

```python
china = data[data.country =='China']
china
```

```python
plt.plot(us.year,us.population / 10**6) # dividing the population by one million ie 10^6.
plt.plot(china.year,china.population / 10**6)
plt.legend(['United States','China'])
plt.xlabel('population')
plt.ylabel('year')
plt.show()
```
